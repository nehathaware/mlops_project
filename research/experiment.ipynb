```python
X = df.drop(['quality'], axis=1)
Y = df["quality"]
X.head()
```

```python
Y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
X_train
```

```python
Y_train
```

```python
X_test
```

```python
from sklearn.linear_model import ElasticNet
lr = ElasticNet(alpha=0.1, l1_ratio=0.5, random_state=42)
lr.fit(X_train,Y_train)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np
def eval_metrics(actual_val, predicted_val):
    rmse = np.sqrt(mean_squared_error(actual_val, predicted_val))
    mae = mean_absolute_error(actual_val, predicted_val)
    r2 = r2_score(actual_val, predicted_val)
    return rmse, mae, r2


predicted_val = lr.predict(X_test)
rmse, mae, r2 = eval_metrics(Y_test, predicted_val)
print("rmse",rmse)
print("mae",mae)
print("r2",r2)
```
